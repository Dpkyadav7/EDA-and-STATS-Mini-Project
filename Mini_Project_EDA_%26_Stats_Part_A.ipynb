```python
'''Q16. Compute a 95% Confidence Interval for the true price of a shirt in the NuCloth website using appropriate distribution.( State reasons as to why did 
you use a z or t distribution)'''

from scipy.stats import norm

sample_mean = np.mean(Data)
sample_std = np.std(Data)

n = len(Data)
sem = sample_std / np.sqrt(n)


z_crit = norm.ppf(0.975) 
me = z_crit * sem

ci_low = sample_mean - me
ci_high = sample_mean + me


print(f"95% Confidence Interval: ({ci_low:.2f}, {ci_high:.2f})")
```

We use a z-distribution to calculate the margin of error, since we have a large sample size (assumed to be at least 30) and do not know the population standard deviation.

```python
'''Q17. A data scientist wants to estimate with 95% confidence the true 
proportion of shirts having price greater than 25 dollars in the NuCloth 
website. A recent study showed that 60% of all shirts have a price greater 
than 25 dollars. The data scientist wants to be accurate within 2% of the true 
proportion. Find the minimum sample size necessary.'''

p_hat = 0.6
d = 0.02
z_alpha = norm.ppf(0.975)  


n = ((z_alpha**2) * p_hat * (1 - p_hat)) / (d**2)
n = np.ceil(n)

print(f"The minimum sample size required is {int(n)}")
```

```python
'''Q18. The same data scientist wants to estimate the true proportion of shirts having price greater than 25 dollars. 
She wants to be 90% confident and accurate within 5% of true proportion. Find the minimum sample size necessary'''
p_hat = 0.6
d = 0.05
z_alpha = norm.ppf(0.95)
n = ((z_alpha**2) * p_hat * (1 - p_hat)) / (d**2)
n = np.ceil(n)

print(f"The minimum sample size required is {int(n)}")
```

```python
'''Q20. A data scientist is researching the hypothesis that the average price of a shirt in NuCloth is higher than the supermarket. So he collects data from NuCloth and the 
supermarket that the average shirt price of shirts in NuCloth is 36 dollars vs 34 dollars in the supermarket. The standard deviations are 7.6 and 5.8 
respectively.Suppose the data scientist got these values after randomly selecting 100 shirts from each place.
a.What hypothesis would he use to compare the shirt prices of NuCloth vs Supermarket
b.What are critical values to be used?
c.What statistical test will be used to compare these prices?
d.Complete the test and obtain the P-value.
e.Summarize his conclusion based on the P-value.'''
```

20.a)
Null Hypothesis: The average price of shirts in NuCloth is not significantly higher than the average price of shirts in the Supermarket.
Alternative Hypothesis: The average price of shirts in NuCloth is significantly higher than the average price of shirts in the Supermarket.

```python
#20.b) 
from scipy.stats import norm
z_critical = norm.ppf(0.05/2)
print(f"The Critical value to be used is {z_critical}")
```

20.c) We will use a two-sample z-test to compare the means of the two populations.

```python
#20.d)
import pandas as pd
from scipy.stats import norm

nucloth = pd.Series([36]*100)
nucloth_std = 7.6

supermarket = pd.Series([34]*100)
supermarket_std = 5.8

test_stat = (nucloth.mean() - supermarket.mean()) / ((nucloth_std**2/100 + supermarket_std**2/100)**0.5)
p_value = 2 * (1 - norm.cdf(abs(test_stat)))

margin_error = 1.96 * ((nucloth_std**2/100 + supermarket_std**2/100)**0.5)
conf_int = (nucloth.mean() - supermarket.mean() - margin_error, nucloth.mean() - supermarket.mean() + margin_error)

print("Test Statistic: ", test_stat)
print("P-value: ", p_value)
print("95% Confidence Interval: ", conf_int)
```

20.e The p-value is less than the significance level of 0.05, which suggests that there is strong evidence to reject the null hypothesis. The confidence interval for the difference in means is positive and does not contain zero, indicating that the average price of shirts in NuCloth is significantly higher than the average price of shirts in the Supermarket.
